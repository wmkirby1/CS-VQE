```python
from utils.molecule_tools import get_molecule
import json
```

```python
file = 'molecule_data'
with open('data/'+file+'.json', 'r') as json_file:
    molecule_data = json.load(json_file)

speciesname_list = list(molecule_data.keys())
```

```python
ham_data = {}
for speciesname in speciesname_list:
    print(speciesname)
    molecule = get_molecule(speciesname)
    
    ham   = molecule['hamiltonian']
    for p in ham.keys():
        ham[p] = ham[p].real
        
    uccsd = molecule['uccsdansatz']
    for p in uccsd.keys():
        uccsd[p] = uccsd[p].imag
    
    num_qubits = molecule['num_qubits']
    ham_data[speciesname] = {"ham":ham,
                             "uccsd":uccsd,
                             "num_qubits":num_qubits}
```

```python
data={}
for k in ham_data.keys():
    data[str(k)] = ham_data[k]
     
with open("hamiltonians/ham_data.json", "w") as outfile: 
    json.dump(data, outfile)
```
